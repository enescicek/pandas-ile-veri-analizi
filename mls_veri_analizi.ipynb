```python
soccer = pd.read_csv("/content/drive/My Drive/github_deneme/mls") # VeriSetini Okur.
```

```python
soccer
```

```python
soccer.head(n = 10) # ilk 10 veriyi okur.
```

```python
len(soccer.index) # Kaç tane Veri Olduğunu Bulur(Satır Sayısı)
```

```python
soccer["base_salary"].mean() # Tüm Maaşların Ortalaması Bulur.
```

```python
soccer["base_salary"].max() # En yüksek maaşı bulur.
```

```python
soccer[soccer["guaranteed_compensation"] == soccer["guaranteed_compensation"].max()]["last_name"].iloc[0] # En yüksek tazminata sahip futbolcunun soyadını bulun
```

```python
soccer[soccer["last_name"] == "Gonzalez Pirez" ]["position"].iloc[0] # "Gonzalez Pirez" isimli futbolcunun oynadığı mevkiyi bulur.
```

```python
soccer.groupby("position").mean()["base_salary"] # Mevkilere göre futbolcuları gruplayarak mevkilerde ortalama maaşları bulur.
```

```python
soccer["position"].nunique() # verisetinde kaç farklı mevki olduğunu bulur.
```

```python
soccer["position"].value_counts() # Her mevkide kaç oyuncunun olduğunu hesaplayır
```

```python
soccer["club"].value_counts() # Her takımda kaç kişinin oynadığını hesaplayır.
```

```python
def son_find(last_name):
    if "son" in last_name.lower():
        return True
    return False

soccer[soccer["last_name"].apply(son_find)] # Soyadlarının içinde "son" geçen futbolcuları bulur
```
